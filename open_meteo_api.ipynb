Use the Weather Forecast API to get the predicted Temperature in your city for the next 7 days.

```python
import requests

response = requests.get(
            'https://api.open-meteo.com/v1/forecast?latitude=45.2662&longitude=11.7721&hourly=temperature_2m&timezone=Europe%2FBerlin')
```

```python
data = response.json()

timestamp = data['hourly']['time']
temperature = data['hourly']['temperature_2m']
print(timestamp)
print(temperature)
```

```python
import matplotlib.pyplot as plt
from datetime import datetime

dates = [datetime.fromisoformat(t) for t in timestamp]

plt.figure(figsize=(12, 5))
plt.plot(dates, temperature, marker='o', markersize=3)
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.title('Hourly Weather Data')

# show ~10 x-tick labels to avoid overcrowding
interval = max(1, len(dates) // 10)
plt.xticks(dates[::interval], rotation=45, ha='right')
plt.tight_layout()
plt.show()

 # nota: matplotlib sceglie automaticamente di non mostrare le ore ma solo le date, 
 # se si vuole mostrare anche l'ora bisogna modificare il formato dei tick sull'asse x, 
 # ad esempio usando plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
```
