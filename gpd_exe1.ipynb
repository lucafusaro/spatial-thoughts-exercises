```python
import geopandas as gpd
```

```python
import os
data_folder_path = 'data'
filename = 'karnataka.gpkg'
path = os.path.join(data_folder_path, filename)
```

```python
roads_gdf = gpd.read_file(path, layer='karnataka_major_roads')
roads_gdf
```

```python
roads_gdf.geometry
```

```python
filtered = roads_gdf[roads_gdf['ref'].str.match('^NH') == True]
filtered
```

```python
filtered.crs
```

```python
roads_reprojected = filtered.to_crs('EPSG:32643')
roads_reprojected.crs
```

```python
roads_reprojected['length'] = roads_reprojected['geometry'].length
```

```python
total_length = roads_reprojected['length'].sum() #meter
print(f'Total length of national highways in the state is {total_length/1000} KM')
```

```python
districts_gdf = gpd.read_file(path, layer='karnataka_districts')
districts_gdf
```

```python
districts_reprojected = districts_gdf.to_crs('EPSG:32643')
```

A spatial join is performed using the sjoin() method. It takes 2 core arguments.

predicate: The spatial predicate to decide which objects to join. Options are intersects (geometries with at least one common point), within (first geometry completely contained in the second one) and contains (second geometry completely contained in the first one).

how: The type of join to perform. Options are left, right and inner.

```python
joined = gpd.sjoin(roads_reprojected, districts_reprojected, how='left', predicate='intersects')
joined
```

In this example, some road segments cross polygon boundaries. A spatial join will duplicate these segments for each polygon they intersect, resulting in an overestimation of the total length. A more accurate method is to use a Spatial Overlay, which splits segments at polygon boundaries. We can implement the overlay as below.

```python
joined = roads_reprojected.overlay(districts_reprojected, how='intersection', keep_geom_type=True)
joined
```

```python
joined['length'] = joined.geometry.length
joined
```

```python
results = joined.groupby('DISTRICT')['length'].sum()/1000
results
```
